```python
# Lists — ordered collections
scores = [85, 92, 78, 96, 88]
print("All scores:", scores)
print("First score:", scores[0])
print("Average:", sum(scores) / len(scores))

# Dictionaries — key-value pairs (like a lookup table)
student = {
    "name": "Astin",
    "degree": "MSc AI & Data Science",
    "university": "Ulster University",
    "graduation_year": 2025
}

print(student["name"])
print(student["degree"])
```

```python
def summarise_student(student_dict):
  name = student_dict["name"]
  degree = student_dict["degree"]
  scores = student_dict["scores"]
  average = sum(scores) / len(scores)

  print(f"{name} studied {degree}")
  print(f"Their average score was: {average:.1f}%")

my_profile = {
    "name": "Astin",
    "degree":  "Msc AI & Data Science",
    "scores": [85, 92, 78, 96, 88]
}

summarise_student(my_profile)
```
